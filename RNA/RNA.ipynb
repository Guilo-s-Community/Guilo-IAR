```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.neural_network import MLPRegressor
from sklearn.preprocessing import MaxAbsScaler

def run_simulation(hidden_layer_sizes, iteracoes=400, n_runs=10):
    errors = []
    for i in range(n_runs):
        print(f"Rodada {i+1} com a arquitetura {hidden_layer_sizes}")

        # Carregando os dados
        arquivo = np.load('teste(1,2,3,4,5).npy')
        x = arquivo[0]
        scale = MaxAbsScaler().fit(arquivo[1])
        y = np.ravel(scale.transform(arquivo[1]))

        # Definindo o modelo
        regr = MLPRegressor(
            hidden_layer_sizes=hidden_layer_sizes,
            max_iter=iteracoes,
            activation='tanh',
            solver='adam',
            learning_rate='adaptive',
            n_iter_no_change=iteracoes,
            verbose=False
        )

        # Treinando o modelo
        regr.fit(x, y)

        # Predição e cálculo do erro
        y_est = regr.predict(x)
        error = np.mean((y - y_est) ** 2)
        errors.append(error)

    # Calculando a média e o desvio padrão do erro
    mean_error = np.mean(errors)
    std_error = np.std(errors)
    return mean_error, std_error, regr

# Definindo as diferentes arquiteturas de rede
architectures = [
    (10, 5),  # 1 camada de 10 neurônios e 1 camada de 5 neurônios
    (15, 10), # 1 camada de 15 neurônios e 1 camada de 10 neurônios
    (20, 15), # 1 camada de 20 neurônios e 1 camada de 15 neurônios
]

# Rodando as simulações para cada arquitetura
results = {}
for arch in architectures:
    print(f"Simulação para a arquitetura {arch}")
    mean_error, std_error, best_regr = run_simulation(arch)
    results[arch] = (mean_error, std_error, best_regr)

# Imprimindo os resultados
for arch, (mean_error, std_error, _) in results.items():
    print(f"Arquitetura: {arch}, Média do Erro: {mean_error:.4f}, Desvio Padrão: {std_error:.4f}")

# Gráficos para o melhor resultado de cada arquitetura
plt.figure(figsize=[14, 7])

# Plotando o melhor resultado
best_arch = min(results, key=lambda arch: results[arch][0])  # Melhor arquitetura com base na média do erro
mean_error, std_error, best_regr = results[best_arch]

# Carregando os dados novamente para o gráfico
arquivo = np.load('teste5.npy')
x = arquivo[0]
scale = MaxAbsScaler().fit(arquivo[1])
y = np.ravel(scale.transform(arquivo[1]))

# Predição do melhor modelo
y_est = best_regr.predict(x)

# Plotando os gráficos
plt.subplot(1, 3, 1)
plt.title('Função Original')
plt.plot(x, y, color='green')

plt.subplot(1, 3, 2)
plt.title(f'Curva erro (Média: {mean_error:.4f}, Desvio Padrão: {std_error:.4f})')
plt.plot(best_regr.loss_curve_, color='red')

plt.subplot(1, 3, 3)
plt.title('Função Original x Função Aproximada')
plt.plot(x, y, linewidth=1, color='green')
plt.plot(x, y_est, linewidth=2, color='blue')

plt.show()
```
